```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
from helpers import (divide_graph_into_parts, calculate_cell_weights, 
                     plot_cell_weights, print_weight_summary, get_grid_dimensions)
```

## Initial Data

```python

# --- DATA ---

# POIs: (x, y, density)
points_of_interest = [
    (1,  15, 0.9),
    (2,  8,  0.6),
    (2,  6,  0.4),
    (7,  12, 0.8),
    (13, 6,  0.3),
]

# Existing charging stations: (x, y)
existing_charging_points = [
    (3,  7),
    (8,  3),
    (11, 8),
]

# Gas stations: (x, y)
gas_stations = [
    (5,  10),
    (10, 5),
    (3,  14),
    (9,  15),
]
```

## Plotting the initial data

```python

# --- PLOT ---

fig, ax = plt.subplots(figsize=(8, 5))

# Colormap for POI density
cmap = cm.YlOrRd
norm = mcolors.Normalize(vmin=0, vmax=1)

# POIs — color mapped to density
poi_x = [p[0] for p in points_of_interest]
poi_y = [p[1] for p in points_of_interest]
poi_d = [p[2] for p in points_of_interest]

sc = ax.scatter(poi_x, poi_y, c=poi_d, cmap=cmap, norm=norm,
                s=150, zorder=5, edgecolors='black', linewidths=0.5)

# Colorbar
cbar = plt.colorbar(sc, ax=ax, pad=0.02, shrink=0.5)
cbar.set_label('Population Density', fontsize=11)

# Existing chargers
ex_x, ex_y = zip(*existing_charging_points)
ax.scatter(ex_x, ex_y, color='red', s=180, marker='*', zorder=5, edgecolors='black', linewidths=0.5)

# Gas stations
gs_x, gs_y = zip(*gas_stations)
ax.scatter(gs_x, gs_y, color='limegreen', s=100, marker='s', zorder=5, edgecolors='black', linewidths=0.5)

# Legend (excluding POIs since colorbar handles that)
ax.scatter([], [], c='orange', s=100, edgecolors='black', linewidths=0.5, label='POI')
ax.scatter([], [], color='red', s=150, marker='*', edgecolors='black', linewidths=0.5, label='Existing Charger')
ax.scatter([], [], color='limegreen', s=100, marker='s', edgecolors='black', linewidths=0.5, label='Gas Station')

ax.set_xticks(range(0, 15, 1))
ax.set_yticks(range(0, 20, 1))
ax.grid(True, linewidth=0.5, alpha=0.3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('EVCP (5,3,3) — Input Data')
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

## Gridding

```python

num_qubits = 8

# Bounds matching your plot axes (clean integers)
X_MIN, X_MAX = 0, 14
Y_MIN, Y_MAX = 0, 19

grid_details, plot_deets = divide_graph_into_parts(
    x_min=X_MIN, x_max=X_MAX,
    y_min=Y_MIN, y_max=Y_MAX,
    num_qubits=num_qubits,
    points_of_interest=points_of_interest,
    existing_chargers=existing_charging_points,
    gas_stations=gas_stations,
    grid_division="default",
    grid_details_flag=True
)

print(f"\nplot_deets: {plot_deets}")
```

## Cell Weight Calculation 
Parameters (tunable):
- `scale_factor = 5.0` (amplifies top end)
- `min_weight = 0.5` (floor for cells with POIs)

```python
# Calculate cell weights
scale_factor = 5.0  
min_weight = 0.5    


cell_weights = calculate_cell_weights(
    grid_details,
    scale_factor=scale_factor,
    min_weight=min_weight
)

# Display summary
print_weight_summary(cell_weights, grid_details, top_n=15)
```

```python
# Visualize cell weights
plot_cell_weights(
    plot_deets, grid_details, cell_weights,
    points_of_interest=points_of_interest,
    existing_chargers=existing_charging_points,
    gas_stations=gas_stations,
    show_data_points=True,
    show_weight_values=False  
)
```
